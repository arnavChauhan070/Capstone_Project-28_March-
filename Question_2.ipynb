```python
"""
 Scenario 2: AI Career Advisor & Learning Path Generator
 Objective:
Build a multi-step intelligent advisor system using agent workflows.

Problem Statement
Students need guidance on career paths based on their skills.

Your system should:

Analyze student profile
Suggest career roles
Generate learning roadmap
 Agents to Build
Profile Analyzer Agent
Understands student background
Career Recommendation Agent
Suggests suitable roles
Skill Gap Agent
Identifies missing skills
Learning Path Agent
Creates step-by-step roadmap
Tasks to Implement
Sequential workflow (LangGraph preferred)
State passing between agents
Conditional logic (beginner vs intermediate)
Sample Input
“I know Python basics and statistics, and I want to become a Data Scientist”

Expected Output
Suggested roles (e.g., Data Scientist, Analyst)
Skill gaps (ML, SQL, etc.)
Learning roadmap (step-by-step)
Tools (Choose Any)
LangGraph (recommended)
CrewAI
AutoGen
Python logic-based workflow
Microsoft Copilot Studio
Deliverables
Workflow implementation
Output for at least 2 student profiles
Diagram of agent flow
"""

# ============================================
# AI Career Advisor (LangGraph Style)
# ============================================

from typing import TypedDict

# --------------------------------------------
# 1. Define State
# --------------------------------------------

class CareerState(TypedDict):
    profile: str
    level: str
    roles: list
    skill_gaps: list
    roadmap: list


# --------------------------------------------
# 2. Agent 1: Profile Analyzer
# --------------------------------------------

def profile_analyzer(state: CareerState):
    text = state["profile"].lower()

    if "basic" in text or "beginner" in text:
        level = "beginner"
    else:
        level = "intermediate"

    return {"level": level}


# --------------------------------------------
# 3. Agent 2: Career Recommendation
# --------------------------------------------

def career_agent(state: CareerState):
    text = state["profile"].lower()

    roles = []

    if "data" in text:
        roles = ["Data Scientist", "Data Analyst"]
    elif "web" in text:
        roles = ["Frontend Developer", "Backend Developer"]
    else:
        roles = ["Software Engineer"]

    return {"roles": roles}


# --------------------------------------------
# 4. Agent 3: Skill Gap Analyzer
# --------------------------------------------

def skill_gap_agent(state: CareerState):

    if state["level"] == "beginner":
        gaps = ["Python", "SQL", "Statistics", "Machine Learning"]
    else:
        gaps = ["Deep Learning", "System Design", "Advanced ML"]

    return {"skill_gaps": gaps}


# --------------------------------------------
# 5. Agent 4: Learning Path Generator
# --------------------------------------------

def learning_path_agent(state: CareerState):

    roadmap = []

    if state["level"] == "beginner":
        roadmap = [
            "Learn Python basics",
            "Study Statistics",
            "Learn SQL",
            "Intro to Machine Learning",
            "Build Projects"
        ]
    else:
        roadmap = [
            "Advanced ML Algorithms",
            "Deep Learning",
            "Deploy Models",
            "System Design",
            "Build Advanced Projects"
        ]

    return {"roadmap": roadmap}


# --------------------------------------------
# 6. Main Workflow (Sequential Execution)
# --------------------------------------------

def career_advisor_system(profile_input):

    state = {
        "profile": profile_input,
        "level": "",
        "roles": [],
        "skill_gaps": [],
        "roadmap": []
    }

    # Step 1
    state.update(profile_analyzer(state))

    # Step 2
    state.update(career_agent(state))

    # Step 3
    state.update(skill_gap_agent(state))

    # Step 4
    state.update(learning_path_agent(state))

    return state


# --------------------------------------------
# 🧪 TEST CASES
# --------------------------------------------

profiles = [
    "I know Python basics and statistics, want to become Data Scientist",
    "I am experienced in web development and want to grow further"
]

for p in profiles:
    result = career_advisor_system(p)

    print("\n PROFILE:", p)
    print(" Level:", result["level"])
    print(" Roles:", result["roles"])
    print(" Skill Gaps:", result["skill_gaps"])
    print(" Roadmap:", result["roadmap"])
```
